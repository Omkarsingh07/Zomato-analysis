Zomato data analysis project

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv("zomato .csv", encoding='latin1')
print(dataframe)
```

Here I change the data type of the rate

```python
def handleRate(value):
    if value == 'NEW' or value == '-':
        return None

    value = str(value).split('/')
    value = value[0]

    return float(value)
dataframe['rate'] = dataframe['rate'].apply(handleRate)
print(dataframe['rate'])
```

```python
dataframe.info()
```

Type of resturant 

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("type of restaurant")
```

conclusion - majority of the customer order from?
answer- delivery.

```python
dataframe.head()
```

```python
group_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result =pd.DataFrame({"votes":group_data})
plt.plot(result,c="black",marker="o")
plt.xlabel("Type of the Restaurant", c="blue",size = 15)
plt.ylabel("votes", c="blue",size=15)
```

conclusion - Delivery restaurant has maximun votes.

```python
plt.hist(dataframe['rate'],bins = 10)
plt.title("rating distribution")
plt.show()
```

conclusion - majority of the resturants received rating in between 3.5 - 4.0

```python
dataframe.head()
```

```python
cost_counts = dataframe['approx_cost(for two people)'].value_counts().head(10)

plt.figure(figsize=(10,5))

sns.barplot(
    x=cost_counts.index,
    y=cost_counts.values
)

plt.xlabel("Approx Cost for Two")
plt.ylabel("Number of Restaurants")

plt.show()
```

```python
cost_counts = dataframe['approx_cost(for two people)'].value_counts()

print(cost_counts)
```

conclusion - average oder expense is 300.

```python
plt.figure(figsize=(6,6))

sns.boxplot(
    x='online_order',
    y='rate',
    data=dataframe
)

plt.show()
```

conclusion - offline rating receive the lower rating in comparison of the online order.

```python
pivot_table = dataframe.pivot_table(
    index='listed_in(type)',
    columns='online_order',
    aggfunc='size',
    fill_value=0
)

sns.heatmap(
    pivot_table,
    annot=True,
    cmap="YlGnBu",
    fmt='d'
)

plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")

plt.show()
```
